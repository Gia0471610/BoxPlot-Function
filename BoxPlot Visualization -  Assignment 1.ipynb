```python
import pandas as pd
import seaborn as sns

from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
```

```python
# Load new cleaned dataset
df = pd.read_csv('googleplaystore_cleaned.csv')
df.head()
```

```python
# Group the data by 'Category' and calculate the median ratings for each category
category_ratings = df.groupby('Category')['Rating'].median().sort_values(ascending=False)
category_ratings
```

```python
# Create a boxplot of the ratings for each category
plt.figure(figsize=(12, 6))
#df.boxplot(column='Rating', by='Category', vert=False, showfliers=False, showmeans=True,
#           meanline=True, meanprops={'color': 'red'}, medianprops={'color': 'blue'})

sns.boxplot(data=df, x='Category', y='Rating', showfliers=False, showmeans=True,
            meanline=True, meanprops={'color': 'red'}, medianprops={'color': 'blue'})

plt.title('Ratings of Apps in Different Categories in Google Play Store')
plt.xlabel('App Category')
plt.ylabel('Rating')
plt.suptitle('')  # Remove default title
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
